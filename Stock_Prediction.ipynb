## BUILDING A RANDOM FOREST MODEL

```python
data_aapl = pd.read_csv('Live Stock Data - aapl_sp_data.csv')
data_aapl_new = data_aapl.drop(columns=['Open','High','Low','Close'], axis=1) 
data_aapl_new.set_index('Date', inplace=True)
```

```python
data_aapl_new
```

```python
# Determine the sentiment threshold
data_aapl_new['Polarity'].hist(bins=30, color='darkred', alpha=0.7)
```

```python
positive_st = 0.13
negative_st = -0.13
```

```python
def sentimenttextblob(polarity):
    if polarity >= positive_st:
        return "Positive"
    elif polarity <= negative_st:
        return "Negative"
    else:
        return "Neutral" 


data_aapl_new['Sentiment'] = data_aapl_new['Polarity'].apply(sentimenttextblob)
data_aapl_new['Sentiment'].value_counts()
```

```python
data_aapl_new['Price_diff'] = data_aapl_new['Adj Close'].diff()
data_aapl_new.dropna(inplace =True)
data_aapl_new.head()
```

```python
Rise = 1
Fall = 0
data_aapl_new['Stock_Trend'] = np.where( data_aapl_new['Price_diff'] > 0, Rise, Fall)
data_aapl_new
```

```python
#Encoding the Sentiments....Binary Encoding the sentiments categories
data_aapl_encoded = data_aapl_new[['Twitter_Volume','Volume','Adj Close','Sentiment','Stock_Trend']]
data_aapl_encoded = pd.get_dummies(data_aapl_encoded, columns=['Sentiment'])
data_aapl_encoded.head()
```

```python
X = data_aapl_encoded.copy()
X.drop('Stock_Trend', axis=1, inplace =True)
X.head()
```

```python
y = data_aapl_encoded['Stock_Trend'].values.reshape(-1,1)
# y = data_aapl_encoded['Stock_Trend']
y[:5]
```

```python
# Spliting the data into x and y
SPLIT = int(.8 * len(X))

X_train = X[:SPLIT]
X_test = X[SPLIT:] 

 
y_train = y[:SPLIT]
y_test = y[SPLIT:]
```

```python
# Scaling the data

scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)

X_test_scaled = X_scaler.transform(X_test)

```

```python
X_train_scaled[:5]
```

# Training a model using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier(n_estimators=200)
rfc_model.fit(X_train_scaled , y_train)
```

```python
#Making predictions on the test set
rfc_model_pred = rfc_model.predict(X_test)

```

# Evaluating the model's performance

```python
#Using accuracy_score
accuracy_rfc_model = accuracy_score(y_test, rfc_model_pred) 
per_accuracy_rfc = (accuracy_rfc_model * 100).round(2)
print( f" The accuracy score of the Decision Tree Model is {accuracy_rfc_model}.")
print( f" In other words {per_accuracy_rfc}% of all the observations are correctly classified.")
```
